```python
import pandas as pd
import numpy as np
import MetaTrader5 as mt5
import datetime as dt
from pandas_datareader import data as pdr
from datetime import datetime, timedelta
```

```python
# initializing connection to MT5 
mt5.initialize()
```

```python
symbols = mt5.symbols_get()

start = dt.datetime.today()

# create an empty list to store historical data
historical_data = []

# loop through symbols and get historical data
for symbol in symbols:
    data = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_D1, start, 30)
    if data is not None:
        data = pd.DataFrame(data)
        data['symbol'] = symbol
        historical_data.append(data)
    else:
        print(f"Skipping symbol {symbol} as no data was returned.")

# concatenate data for each symbol into a single dataframe
prices = pd.concat(historical_data, keys=[i['symbol'] for i in historical_data])
prices
```
