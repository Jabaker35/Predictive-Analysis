```python
yf.pdr_override()

# retrieving the ticker list from S&P 500 w/o typing the ticker symbols
assets = si.tickers_nasdaq()
# financial = YahooFinancials(assets)

# creating date variables going back three years to today's date
three = dt.timedelta(days = 365) * 3
begin = (dt.datetime.today() - dt.timedelta(30)).strftime('%Y-%m-%d')
stop = dt.datetime.today().strftime('%Y-%m-%d')

# # retrieving historical data through Yahoo Financials
# data = financial.get_historical_price_data('2022-12-01', '2023-01-02', 'daily')

# # creating a database for Adjusted Close on every ticker symbol
# prices = pd.DataFrame({
#     a: {x['formatted_date']: x['adjclose'] for x in data[a]['prices']} for a in assets
# })

# # naming index column
# prices.index.name = 'Date'

# creating the dataframe for the adjusted close
prices = pd.DataFrame()
for stock in assets:
    prices[stock] = pdr.DataReader(stock, data_source='yahoo', start = begin, end = stop)['Adj Close']

# displaying the dataframe
prices
```

```python
# drop na or missing values 
prices.dropna(inplace = True, axis = 1)
```

```python
# calculate the Simple Returns
daily_simple_returns = prices.pct_change()
daily_simple_returns
```

```python
# create and show annualized covariance matrix
cov_matrix_annual = daily_simple_returns.cov() * 252
cov_matrix_annual
```

```python
# calculate the mean fo the daily simple returns
daily_mean = daily_simple_returns.mean()
daily_mean
```

```python
# show the annualized returns by multiplying trading days
daily_mean * 252
```

```python
# sort the values for the daily mean simple returns & show annualized returns as a percentage
daily_mean.sort_values(axis = 0, ascending = False, inplace = True, kind='quicksort', na_position='last', ignore_index='False', key=None)
daily_mean * 252 * 100
```

```python
# optimize the portfolio
# calculate the expected returns & the ann sample covariance matrix of (daily) asset returns
mu = expected_returns.mean_historical_return(prices)
S = risk_models.sample_cov(prices)
```

```python
# optimize for maximal Sharpe Ratio
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
ef.portfolio_performance(verbose = True)
print(cleaned_weights)
```

```python
# get the discrete allocation of each share per stock
# portfolio_val = input(int())
latest_prices = prices.tail(1).squeeze()

weights = cleaned_weights
da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=500)
allocation, leftover = da.lp_portfolio()
print('DiscreteAllocation:', allocation)
print('Funds Remaining: ${:.2f}', leftover)
```

```python
# get the discrete allocation values
discrete_list = []
for symbol in allocation:
    discrete_list.append(allocation.get(symbol))
```

```python
yf.pdr_override()

begin = (dt.datetime.today() - dt.timedelta(365)).strftime('%Y-%m-%d')
stop = dt.datetime.today().strftime('%Y-%m-%d')

# choose a stock symbol of your choice based on the previous predictions
ticker = input(str())
df = pdr.get_data_yahoo(ticker, start=begin, end=stop)

# show the data
df
```

```python
# visualize the closing price data
plt.figure(figsize=(16,8))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price USD ($)', fontsize = 18)
plt.show()
```

```python
# create a new dataframe with only the close column
data = df[['Close']]
data.head()
```

```python
# create a variable to predict 'x' days out into the future
future = 25

# create a new column (target) shifted 'x' units/days up
data['Prediction'] = data[['Close']].shift(-future)
data.tail(4)
```

```python
# create the feature data set (x) and convert it to a numpy array and remove the last 'x' rows/days
x = np.array(data.drop(['Prediction'], 1))[: -future]
print(x)
```

```python
# create the target data set (y) and convert it to a numpy array and get all of the target values except the last 'x' days
y = np.array(data['Prediction'])[:-future]
print(y)
```

```python
# split data into 75% training and 25% testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
# create the models
# create the decision tree regressor model
tree = DecisionTreeRegressor().fit(x_train, y_train)

# create the linear regression model
lr = LinearRegression().fit(x_train, y_train)
```

```python
# retrieving the last 'x' rows of the feature data set
x_future = data.drop(['Prediction'], 1)[:-future]
x_future = x_future.tail(future)
x_future = np.array(x_future)
x_future
```

```python
# show the model tree prediction
tree_pred = tree.predict(x_future)
print(tree_pred)
print()
# show the model linear regression prediction
lr_pred = lr.predict(x_future)
print(lr_pred)
```

```python
# visual the data
predictions = tree_pred

valid = data[x.shape[0]:]
valid['Predictions'] = predictions

plt.figure(figsize=(16,8))
plt.title('Valid Model')
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df['Close'])
plt.plot(valid[['Close','Predictions']])
plt.legend(['Original','Valid','Prediction'])

plt.show()
```

```python
# visual the data
predictions = lr_pred

valid = data[x.shape[0]:]
valid['Predictions'] = predictions

plt.figure(figsize=(16,8))
plt.title('Linear Model')
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df['Close'])
plt.plot(valid[['Close','Predictions']])
plt.legend(['Original','Valid','Prediction'])

plt.show()
```
